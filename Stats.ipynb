**Measure of Central Tendency**

- `Mean `
- `Median `
- `Mode `

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'Revenue':[4,3,1,7,11,12,9,6,14,21,9,6,9,80,6]})
df
```

```python
# Mean -> sum of the values / Number of  values

c = 0
d = 0
for i in df['Revenue']:
  c = c + i
  d = d + 1

print(c / d)
```

```python
df['Revenue'].mean()
```

```python
df['Revenue'].median()

```

```python
df['Revenue'].mode()[1]
```

**Percentile**

- `(value below (x)) / n * (100)`

```python
c = 0

for i in df['Revenue']:
  if i == 12:
    break
  c = c + 1


f = (c / len(df)) * 100

print(f)
```

**Outliers**

```python
a = [1,2,2,2,3,3,4,5,5,5,6,6,6,6,7,8,8,9,27]


df = pd.DataFrame({'Age':a})
df
```

```python
import seaborn as sns
```

```python
sns.boxplot(x = df['Age'])
```

```python
# capping

IQR = df['Age'].quantile(0.75) - df['Age'].quantile(0.25)

lower_limit = df['Age'].quantile(0.25) - (1.5 * IQR)
upper_limit = df['Age'].quantile(0.75) + (1.5 * IQR)
```

```python
lower_limit
```

```python
upper_limit
```

```python
df['Age_1'] = np.where(df['Age'] > upper_limit,upper_limit,
         np.where(df['Age'] < lower_limit,lower_limit,df['Age']))
```

```python
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (5,3))
sns.boxplot(x = df['Age'])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (5,3))
sns.boxplot(x = df['Age_1'])
```

```python
rohit_values = [18,45,88,0,4,8]
kohli_values = [27,29,32,36,24,33]

import pandas as pd

df = pd.DataFrame({'Rohit_sharma':rohit_values,'Kohli':kohli_values})
```

```python
df
```

```python
# rohit_sharma_variance -> sumation(xi - mean(x) / n-1)

df['Rohit_sharma'].var()
```

```python
df['Rohit_sharma'].std()
```

```python
c = 0
d = 0

for i in df['Rohit_sharma']:
  c = c + i
  d = d + 1

mean_rohith = c / d
f = 0

for j in df['Rohit_sharma']:
  f = f + (j - mean_rohith) ** 2

print(f / (d - 1))
```

**std**

```python
f = pd.DataFrame({'Age':[14,19,7,2,22,36]})
f
```

```python
f['Age'].mean()
```

```python
f['Age'].std()
```

```python
import seaborn as sns
```

```python
# displot

sns.displot(f['Age'],kind = 'kde')
```

```python
# z_score -> Data Scaling -> xi - mean / std


k = []


for i in f['Age']:
  k.append((i - f['Age'].mean()) / f['Age'].std())

print(k)
```

```python
f['Age-scaled'] = k
```

```python
f
```

```python
sns.displot(f['Age-scaled'],kind = 'kde')
```
